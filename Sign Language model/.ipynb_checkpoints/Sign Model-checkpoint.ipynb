```python
import cv2
import numpy as np
import ast
import pandas as pd
import json
import os
from matplotlib import pyplot as plt
import time
import mediapipe as mp
```

```python
# STEP 1 : DETECTING KEYPOINTS USING HOLISTIC
```

```python
# HOLISTIC IS A LIBRARY OF MODELS JUST LIKE JOBLIB
#HOLISTIC MODEL
mp_holistic = mp.solutions.holistic
# MODEL WHICH WILL HELP US TO DRAW THE BODY COORDINATES/LANDMARKS
mp_drawing = mp.solutions.drawing_utils
```

```python
# HERE IMAGE MEANS A FRAME OF VIDEO IS PASSED
def mediapipe_detection(image,model):
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB) #COLOR CONVERTING BGR TO RGB
    image.flags.writeable = False;
    results = model.process(image);
    image.flags.writeable = True;
    image = cv2.cvtColor(image,cv2.COLOR_RGB2BGR) #COLOR CONVERTING RGB TO BGR
    return image,results
```

```python
def draw_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS)#DRAW FACE CONNECTIONS
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)#DRAW BODY/POSE CONNECTIONS
    mp_drawing.draw_landmarks(image, results.pose_world_landmarks, mp_holistic.POSE_CONNECTIONS)#DRAW BODY/POSE CONNECTIONS
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)#DRAW LEFTHAND CONNECTIONS
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)#DRAW RIGHTHAND CONNECTIONS
    
```

```python
def draw_styled_landmarks(image, results):
    #DRAW FACE CONNECTIONS
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS,
                               mp_drawing.DrawingSpec(color=(80, 110, 10), thickness=1, circle_radius=1),
                               mp_drawing.DrawingSpec(color=(80, 256, 121), thickness=1, circle_radius=1)
                             )
    #DRAW BODY/POSE CONNECTIONS
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                               mp_drawing.DrawingSpec(color=(80, 22, 10), thickness=2, circle_radius=4),
                               mp_drawing.DrawingSpec(color=(80, 44, 121), thickness=2, circle_radius=2)
                             )
    #DRAW BODY/POSE CONNECTIONS
    mp_drawing.draw_landmarks(image, results.pose_world_landmarks, mp_holistic.POSE_CONNECTIONS,
                               mp_drawing.DrawingSpec(color=(22, 255, 0), thickness=2, circle_radius=4),
                               mp_drawing.DrawingSpec(color=(255, 237, 0), thickness=2, circle_radius=2)
                             )
    #DRAW LEFTHAND CONNECTIONS
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                               mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                               mp_drawing.DrawingSpec(color=(121, 44, 250), thickness=2, circle_radius=2)
                             )
    #DRAW RIGHTHAND CONNECTIONS
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                               mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=4),
                               mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2)
                             )
```

```python
# GET CAMERA ACCESS
cap=cv2.VideoCapture(0);
# IMPORT THE MEDIA PIPE MODEL FIRST
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5,)as holistic:
    
# START THE LOOP ONLY If CAM IS OPENED
    while cap.isOpened():
    # READ THE VIDEO OR CAM INPUT(CURRENT FRAME)....READ FRAME
        ret,frame=cap.read()

        # MAKE DETECTIONS
        image, results = mediapipe_detection(frame,holistic) #THIS IS A MEDIAPIPE MODEL
        print(results)

        # DRAW LANDMARKS/CONNECTIONS FROM THE DETECTED COORDINATES/POINTS(RESULTS)
        draw_styled_landmarks(frame,results)
        
        # DISPLAY THE CAPTURED "FRAME" TO SCREEN ....SHOW FRAME
        cv2.imshow('OpenCV Feed' , frame)

        # EXIT THE POP UP WHEN PRESSED q OR AFTER 10 MIN IS OVER
        if cv2.waitKey(10) &0xFF == ord('q'):
            break

    # CLOSE CAM ACCESS AND END FRAMES CAPTURED
    cap.release()
    cv2.destroyAllWindows()
```

```python
# NOW PUTTING ALL THIS IN TO FUNCTION SO THAT ALL BODY KEYPOINTS/LANDMARKS ARE STORED AT THE SAME TIME
def extract_keypoints(results):
    pose_world=np.array([[res.x, res.y, res.z]for res in results.pose_world_landmarks.landmark]).flatten() if results.pose_world_landmarks else np.zeros(33*4)
    face=np.array([[res.x, res.y, res.z]for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    left_hand=np.array([[res.x, res.y, res.z]for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    right_hand=np.array([[res.x, res.y, res.z]for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose,face,left_hand,right_hand])
```

```python
# ABOVE FUNCTION :CONCATENATE CLUBS ALL THE BODY PART NUMPY ARRAY(VESSEL) IN TO ONE AND THIS CONCATENATED SET WILL BE USED TO DETECT
```

```python
# PATH TO STORE EXTRACTED LANDMARKS
DATA_PATH = os.path.join('MP_Data')

# WORDS FOR DETECTION
actions = np.array([])  # Initialize an empty NumPy array

with open(r'C:\Users\arpsf\OneDrive\Desktop\Major project 2024\Data\WLASL_v0.3.json', 'r') as file:
    data = json.load(file)

for obj in data:
    actions = np.append(actions, obj["gloss"])
```

```python
len(actions)
```

```python
# NO OF VIDEOS FOR TRAINING
no_sequences = 30
# NO OF FRAMES IN EACH VIDEO
length_sequence = 30
```

```python
for action in actions:
    for sequence in range(no_sequences):
        try:
            os.makedirs(os.path.join(DATA_PATH,action,str(sequence)))
        except: 
            pass
```

```python
# GET CAMERA ACCESS
cap=cv2.VideoCapture(0);
# IMPORT THE MEDIA PIPE MODEL FIRST
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5,)as holistic:
    
    # LOOPING ON ACTIONS(new)
    for action in action:
        for sequence in range(no_sequences):
            # LOOP ON FRAMES Of VIDEOS
            for frame_num in range(length_sequence):
                
                # READ THE VIDEO OR CAM INPUT(CURRENT FRAME)....READ FRAME
                ret,frame=cap.read()
                        
                # MAKE DETECTIONS
                image, results = mediapipe_detection(frame,holistic) #THIS IS A MEDIAPIPE MODEL
                print(results)
                    
                # DRAW LANDMARKS/CONNECTIONS FROM THE DETECTED COORDINATES/POINTS(RESULTS)
                draw_styled_landmarks(frame,results)
    
                # (NEW)collecting frames and wait logic
                if frame_num==0:
                    cv2.putText(image, 'STARTING COLLECTION', (120,200),
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 4, cv2.LINE_AA)
                    cv2.putText(image, 'Collecting Frames for {} Video Number {}'.format(action,sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,225), 1, cv2.LINE_AA)
                    # show to screen
                    cv2.imshow('SMILE PLEASE!!!', frame)
                    cv2.waitKey(2000)
                else:
                    cv2.putText(image, 'Collecting Frames for {} Video Number {}'.format(action,sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,225), 1, cv2.LINE_AA)
                        
                    # DISPLAY THE CAPTURED "FRAME" TO SCREEN ....SHOW FRAME
                    cv2.imshow('SMILE PLEASE!!!' , frame)
                    
                # EXTRACT KEYPOINTS FROM TRAINING VIDEOS AND SAVE
                keypoints = extract_keypoints(results)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                np.save(npy_path, keypoints)
                    
    
                # EXIT THE POP UP WHEN PRESSED q AND FREEZE THE FRAME ON WINDOW TILL THEN
                if cv2.waitKey(10) &0xFF == ord('q'):
                    break

    # CLOSE CAM ACCESS AND END FRAMES CAPTURED
    cap.release()
    cv2.destroyAllWindows()
```
